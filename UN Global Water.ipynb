# UN Global Water Data 2012-2022

About Dataset
Dataset Overview:
The "UN Global Water Data 2012-2022" dataset compiles global water access and sanitation metrics, highlighting changes over a decade.

Column Descriptors:

DATAFLOW: Data collection framework
REF_AREA: Geographic area covered
INDICATOR: Specific metric measured
SEX: Gender breakdown, if applicable
TIME_PERIOD: Year of data collection
OBS_VALUE: The observed value or measurement
UNIT_MEASURE: Units of measurement used
Additional columns include details on data sources, observation status, and methodological notes.
Ethically Collected Data:
Ensures adherence to ethical standards in data collection, respecting privacy and consent.

## Dropping multiple columns

```python
dfNEW=df[['DATAFLOW', 'REF_AREA:Geographic area', 'INDICATOR:Indicator',
       'SEX:Sex', 'TIME_PERIOD:Time period', 'OBS_VALUE:Observation Value',
       #'UNIT_MULTIPLIER:Unit multiplier', 
        'UNIT_MEASURE:Unit of measure',
       #'OBS_STATUS:Observation Status', 'OBS_CONF:Observation confidentaility',
       #'LOWER_BOUND:Lower Bound', 'UPPER_BOUND:Upper Bound',
       #'WGTD_SAMPL_SIZE:Weighted Sample Size',
       #'OBS_FOOTNOTE:Observation footnote', 'SERIES_FOOTNOTE:Series footnote',
       'DATA_SOURCE:Data Source',
       #'SOURCE_LINK:Citation of or link to the data source',
       #'CUSTODIAN:Custodian',
       #'TIME_PERIOD_METHOD:Time period activity related to when the data are collected',
       #'REF_PERIOD:Reference Period',
       #'COVERAGE_TIME:The period of time for which data are provided',
       'AGE:Current age'
]].copy()
```

```python
dfNEW.shape
```

## Remove all the rows that contain a missing value

```python
# remove all the rows that contain a missing value
dfNEW=dfNEW.dropna()
```

## Drops all duplicates

```python
# we still need to drop some columns
df_NEW=dfNEW[['DATAFLOW', 'REF_AREA:Geographic area', 'INDICATOR:Indicator',
       #'SEX:Sex', 
        'TIME_PERIOD:Time period', 'OBS_VALUE:Observation Value',
        'UNIT_MEASURE:Unit of measure',
       'DATA_SOURCE:Data Source',
       #'AGE:Current age'
]]
df_NEW.head()
```

# EDA of the Cleaned data

## Countries Observation Value

```python
REF_AREA=(df_NEW
 .groupby(['REF_AREA:Geographic area'])
 .agg(SUM_ESTIMATE_OBS_VALUE=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_OBS_VALUE'],ascending=False)
REF_AREA
```

## Data source analysis

```python
DATA_SOURCE=(df_NEW
 .groupby(['DATA_SOURCE:Data Source'])
 .agg(SUM_ESTIMATE_OBS_VALUE=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_OBS_VALUE'],ascending=False)
DATA_SOURCE
```

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
plt.style.use("ggplot")
plt.figure(figsize=(10, 15))
DATA_SOURCE.plot.barh(stacked=True, color='r',alpha=0.5)
plt.show()
```

## Observation Value analysis of the cleaned data

```python
OBS_VALUE=(df_NEW
 .groupby(['TIME_PERIOD:Time period'])
 .agg(SUM_ESTIMATE_OBS_VALUE=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_OBS_VALUE'],ascending=False)
OBS_VALUE
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(6, 10))
OBS_VALUE.plot.bar(stacked=True, color='r',alpha=0.5)
plt.show()
```

## Indicator Value of the cleaned data

```python
INDICATOR=(df_NEW
 .groupby(['INDICATOR:Indicator'])
 .agg(SUM_ESTIMATE_INDICATOR=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_INDICATOR'],ascending=False)
INDICATOR
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(15,20))
INDICATOR.plot.barh(stacked=True, color='r',alpha=0.5)
plt.show()
```

# Analysing the data from two west African countries to determine which country has the highest proportion of population using improved drinking water sources

# Geographic analysis of Nigeria

```python
# let's check the Country Name in line with this parameters:'INDICATOR:Indicator','TIME_PERIOD:Time period','OBS_VALUE:Observation Value','DATA_SOURCE:Data Source'
def Country_Name(x):
    return df_NEW[(df_NEW['REF_AREA:Geographic area'] == x)][['INDICATOR:Indicator','TIME_PERIOD:Time period','OBS_VALUE:Observation Value','DATA_SOURCE:Data Source']]
Nigerian_Region=Country_Name("NGA: Nigeria")
Nigerian_Region
```

```python
Nigeria_OBS_VALUE=(Nigerian_Region
 .groupby(['TIME_PERIOD:Time period'])
 .agg(SUM_ESTIMATE_Nigeria_OBS_VALUE=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_Nigeria_OBS_VALUE'],ascending=False)
Nigeria_OBS_VALUE
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(6, 10))
Nigeria_OBS_VALUE.plot.bar(stacked=True, color='r',alpha=0.5)
plt.show()
```

```python
Nigeria_INDICATOR=(Nigerian_Region
 .groupby(['INDICATOR:Indicator'])
 .agg(SUM_ESTIMATE_Nigeria_INDICATOR=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_Nigeria_INDICATOR'],ascending=False)
Nigeria_INDICATOR
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(15,20))
Nigeria_INDICATOR.plot.barh(stacked=True, color='r',alpha=0.5)
plt.show()
```

```python
Nigeria_DATA_SOURCE=(Nigerian_Region
 .groupby(['DATA_SOURCE:Data Source'])
 .agg(SUM_ESTIMATE_Nigeria_DATA_SOURCE=('OBS_VALUE:Observation Value','sum'))
).sort_values(by=['SUM_ESTIMATE_Nigeria_DATA_SOURCE'],ascending=False)
Nigeria_DATA_SOURCE
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(10, 15))
Nigeria_DATA_SOURCE.plot.barh(stacked=True, color='r',alpha=0.5)
plt.show()
```
